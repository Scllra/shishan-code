```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_plot = np.linspace(-10, 10, 400)
y_plot = torch.tensor((x_plot + 1) ** 2 * (x_plot - 1) ** 2)

# 绘制曲线
plt.plot(x_plot, y_plot)

plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.title('Curve of Function f(x) = (x+1)^2*(x-1)^2')
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

x0 = torch.tensor(5.0, requires_grad=True)
y0 = (x0+1)**2*(x0-1)**2

alpha = 0.001
epsilon = 0.0001

x_values = []
y_values = []

# 循环直到y的变化小于epsilon
while True:
    # 计算梯度
    y0.backward()
    y = y0.item()  # 保存y0在当前迭代的取值，用于判断迭代的停止条件
    # 更新x0
    with torch.no_grad():
        x0 -= alpha * x0.grad  # x0 = x0-alpha * x0.grad 创建一个新的张量并赋值给x
    x0.grad.zero_()
    # 计算新的y0
    y0 = (x0+1)**2*(x0-1)**2
    # 检查是否满足结束条件
    if abs(y0.item() - y) < epsilon:
        break

y0.backward()
print("x0: ", x0, "y0: ", y0, "x0.grad: ", x0.grad)
```
